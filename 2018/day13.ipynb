>     \----/
>
> Intersections occur when two perpendicular paths cross. At an intersection, a cart is capable of turning left, turning right, or continuing straight. Here are two loops connected by two intersections:
> 
>     /-----\
>     |     |
>     |  /--+--\
>     |  |  |  |
>     \--+--/  |

>

This process is independent of the particular intersection at which the cart has arrived - that is, the cart has no per-intersection memory.

They do this based on their current location: carts on the top row move first (acting from left to right), then carts on the second row move (again from left to right), then carts on the third row, and so on.
```python
import numpy as np

UP, DOWN, LEFT, RIGHT = '^', 'v', '<', '>'
CAR_CHARACTERS = [UP, DOWN, LEFT, RIGHT]
TURNS = ['/', '\\', '+']
LEFT_TURN, STRAIGHT, RIGHT_TURN = 0, 1, 2
TURN_MAP = {
    # (track char, direction): (new direction)
    ('/', UP): RIGHT,
    ('/', DOWN): LEFT,
    ('/', LEFT): DOWN,
    ('/', RIGHT): UP,
    ('\\', UP): LEFT,
    ('\\', DOWN): RIGHT,
    ('\\', LEFT): UP,
    ('\\', RIGHT): DOWN,
    (LEFT_TURN, UP): LEFT,
    (LEFT_TURN, DOWN): RIGHT,
    (LEFT_TURN, LEFT): DOWN,
    (LEFT_TURN, RIGHT): UP,
    (STRAIGHT, UP): UP,
    (STRAIGHT, DOWN): DOWN,
    (STRAIGHT, LEFT): LEFT,
    (STRAIGHT, RIGHT): RIGHT,
    (RIGHT_TURN, UP): RIGHT,
    (RIGHT_TURN, DOWN): LEFT,
    (RIGHT_TURN, LEFT): UP,
    (RIGHT_TURN, RIGHT): DOWN,
}
```

```python
def array_iter(a):
    I, J = a.shape
    for i in range(I):
        for j in range(J):
            yield i, j, a[i, j]
```

```python
class Car(object):
    def __init__(self, i, j, direction):
        self.i = i
        self.j = j
        self.direction = direction
        self.turn_type = LEFT_TURN
    
    def step(self):
        if self.direction == UP:
            self.i -= 1
        elif self.direction == DOWN:
            self.i += 1
        elif self.direction == LEFT:
            self.j -= 1
        elif self.direction == RIGHT:
            self.j += 1
            
        return self.i, self.j
    
    def turn(self, track_char):
        if track_char in ['/', '\\']:
            self.direction = TURN_MAP[track_char, self.direction]
        elif track_char == '+':
            self.direction = TURN_MAP[self.turn_type, self.direction]
            self.turn_type += 1
            self.turn_type %= 3
```

```python
import collections

class CrashException(Exception):
    pass


class Race(object):
    def __init__(self, input_str):
        self.input_str = input_str
        self.array = np.array([
            list(row) for row in self.input_str.split('\n') if row.strip()
        ])
        self.init_track()
        self.init_cars()
        self.num_steps = 0
        self.log_track()
    
    def init_track(self):
        """build a representation of the underlying track as if no cars 
        were there. normally we'd have to check neighbors to confirm
        track direction, but at the start of the race every car is on a
        straightaway (no cars at corners or crossings)
        
        """
        self.track = self.array.copy()
        
        for i, j, c in array_iter(self.track):
            if c in ['<', '>']:
                self.track[i, j] = '-'
            elif c in ['^', 'v']:
                self.track[i, j] = '|'
    
    def init_cars(self):
        self.cars = []
        for (i, j, c) in array_iter(self.array):
            if c in CAR_CHARACTERS:
                self.cars.append(Car(i, j, c))
        
    def run_until_crash(self):
        while True:
            crash_location = self.step()
            if crash_location is not None:
                return crash_location
    
    def step(self):
        for c in self.cars:
            new_i, new_j = c.step()
            
            # after every move, check for other cars at that location
            if self.crash:
                return new_i, new_j
            
            # now turn when needed
            track_char = self.track[new_i, new_j]
            if track_char in TURNS:
                c.turn(track_char)
        
        self.num_steps += 1
        
        self.log_track()
        
        # no crash
        return None
    
    def log_track(self):
        LOGGER.debug('track after step {}:\n{}'.format(self.num_steps, self.str))
    
    @property
    def car_positions(self):
        return collections.Counter((c.i, c.j) for c in self.cars)
    
    @property
    def crash(self):
        return max(self.car_positions.values()) > 1
    
    @property
    def str(self):
        z = self.track.copy()
        for c in self.cars:
            z[c.i, c.j] = c.direction
        s = '\n'.join([''.join(row) for row in z])
        return s
```

```python
def q_1(input_str):
    t = Race(input_str)
    return t.run_until_crash()
```

```python
from collections import defaultdict
from typing import List, Tuple, Dict


class Cart:
    def __init__(self, pos: complex, di: complex):
        self.position = pos
        self.direction = di
        self.cross_mod = 0
        self.dead = False


def setup(input_file_lines: List[str]) -> Tuple[Dict[complex, str], List[Cart]]:
    tracks = defaultdict(lambda: "")  # only stores important tracks: \ / +
    carts = []
    for y, line in enumerate(input_file_lines):
        for x, char in enumerate(line):
            if char == "\n":
                continue
            if char in "<v>^":
                direction = {
                    "<": -1,
                    "v": +1j,
                    ">": +1,
                    "^": -1j,
                }[char]
                carts.append(Cart(x + y * 1j, direction))  # location, direction, crossings
                part = {
                    "<": "-",
                    "v": "|",
                    ">": "-",
                    "^": "|",
                }[char]
            else:
                part = char
            if part in "\\/+":
                tracks[(x + y * 1j)] = part
    return tracks, carts


def turn_cart(cart: Cart, part: str):
    """This space uses a downwards-facing Y axis, which means all calculations
    must flip their imaginary part. For example, rotation to the left
    (counterclockwise) would be multiplying by -1j instead of by +1j."""
    if not part:  # empty track is impossible, and | or - don't matter
        return
    if part == "\\":
        if cart.direction.real == 0:
            cart.direction *= -1j  # ⮡ ⮢
        else:
            cart.direction *= +1j  # ⮧ ⮤
    if part == "/":
        if cart.direction.real == 0:
            cart.direction *= +1j  # ⮣ ⮠
        else:
            cart.direction *= -1j  # ⮥ ⮦
    if part == "+":
        cart.direction *= -1j * 1j ** cart.cross_mod  # rotate left, forward, or right
        cart.cross_mod = (cart.cross_mod + 1) % 3


def solve_a(input_file_lines: List[str]) -> str:
    tracks, carts = setup(input_file_lines)
    step = 0
    while True:
        print(step)
        step += 1
        carts.sort(key=lambda c: (c.position.imag, c.position.real))
        for ci, cart in enumerate(carts):
            cart.position += cart.direction
            if any(c2.position == cart.position for c2i, c2 in enumerate(carts) if c2i != ci):
                return str(int(cart.position.real)) + "," + str(int(cart.position.imag))
                # 14, 42
            part = tracks[cart.position]
            turn_cart(cart, part)


def solve_b(input_file_lines: List[str]) -> str:
    tracks, carts = setup(input_file_lines)
    while len(carts) > 1:
        carts.sort(key=lambda c: (c.position.imag, c.position.real))
        for ci, cart in enumerate(carts):
            if cart.dead:
                continue
            cart.position += cart.direction
            for ci2, cart2 in enumerate(carts):
                if ci != ci2 and cart.position == cart2.position and not cart2.dead:
                    cart.dead = True
                    cart2.dead = True
                    break
            if cart.dead:
                continue
            part = tracks[cart.position]
            turn_cart(cart, part)
        carts = [c for c in carts if not c.dead]
    if not carts:
        return "ERROR: there's an even number of carts, there's isn't 1 cart left at the end!"
    cart = carts[0]
    return str(int(cart.position.real)) + "," + str(int(cart.position.imag))
    # 8,7
```

```python
solve_a(load_data().split('\n'))
```
