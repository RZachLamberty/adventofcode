```python
def load_data(fname=FNAME):
    with open(fname) as fp:
        return [line.strip() for line in fp]
```

```python
import re

def parse_pv(line):
    return [
        int(_) 
        for _ in re.match(
            'position=< *([\-\d]+), *([\-\d]+)> velocity=< *([\-\d]+), *([\-\d]+)>',
            line
        ).groups()]
```

#### function def

```python
# I'm using py3.6 booooooooo
# @dataclass
# class Point:
#     x: int
#     y: int
#     v_x: int
#     v_y: int
#     t: int

#     def step(self):
#         self.x += self.v_x
#         self.y += self.v_y
#         self.t += 1

import functools

import numpy as np
import tqdm

from ipywidgets import interact


class Point(object):
    t_max = None
    
    def __init__(self, x, y, v_x, v_y, t=0, t_max=100):
        self.x = x
        self.y = y
        self.v_x = v_x
        self.v_y = v_y
        self.t = t
        t_max = t_max
    
    def step(self):
        self.x += self.v_x
        self.y += self.v_y
        self.t += 1
        

class Points(object):
    def __init__(self, pvs, t=0):
        self.t = t
        self.points = [Point(x, y, v_x, v_y, t) for (x, y, v_x, v_y) in pvs]
        self.history = {}
        self.update_history()
        
    def update_history(self):
        self.history[self.t] = {(pt.x, pt.y) for pt in self.points}
    
    def step(self):
        for pt in self.points:
            pt.step()
        self.t += 1
        self.update_history()
    
    def field(self, t=0):
        """return the min and max values in all of self.history"""
        min_x = min(x for (x, y) in self.history[t])
        max_x = max(x for (x, y) in self.history[t])
        min_y = min(y for (x, y) in self.history[t])
        max_y = max(y for (x, y) in self.history[t])
        return min_x, max_x, min_y, max_y
    
    def array_str(self, t=None, bounds=None):
        t = t or self.t
        
        assert t >= 0
        while t not in self.history:
            self.step()
        
        min_x, max_x, min_y, max_y = self.field(t)
        
        if bounds:
            max_width, max_height = bounds
            if max_y - min_y > max_height or max_x - min_x > max_width:
                return 'too big to print'
        
        s = ''
        for y in range(min_y, max_y + 1):
            for x in range(min_x, max_x + 1):
                s += '#' if (x, y) in self.history[t] else '.'
            s += '\n'
        
        return s
    
    def show(self, t=None, bounds=None):
        print(self.array_str(t, bounds))
    
    def printables(self, tmin=0, tmax=100, bounds=None):
        """return a generator of printable array strings"""
        for t in tqdm.tnrange(tmin, tmax + 1):
            s = self.array_str(t, bounds)
            if s != 'too big to print':
                yield t, s
```

```python
pts = Points([parse_pv(line) for line in test_data])

@interact(t=(0, 100, 1))
def show(t=0):
    return pts.show(t)
```
