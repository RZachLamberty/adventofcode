## part 1

### problem statement:

#### loading data

```python
test_data = """987654321111111
811111111111119
234234234234278
818181911112111"""
```

```python
def load_data(fname=FNAME):
    with open(fname) as fp:
        return fp.read()
```

```python
def parse_raw_data(data: str) -> list[list[int]]:
    return [[int(_) for _ in row] for row in data.strip().split('\n')]
```

```python
parse_raw_data(test_data)
```

#### function def

```python
def get_joltage(row: list[int]) -> int:
    n0 = max(row[:-1])
    i0 = row.index(n0)
    n1 = max(row[i0 + 1:])
    return 10 * n0 + n1

assert get_joltage(parse_raw_data(data="987654321111111")[0]) == 98
assert get_joltage(parse_raw_data(data="811111111111119")[0]) == 89
assert get_joltage(parse_raw_data(data="234234234234278")[0]) == 78
assert get_joltage(parse_raw_data(data="818181911112111")[0]) == 92
```

```python
def q_1(data):
    return sum(get_joltage(row) for row in parse_raw_data(data=data))
```

#### tests

```python
def test_q_1():
    LOGGER.setLevel(logging.DEBUG)
    assert q_1(test_data) == 357
    LOGGER.setLevel(logging.INFO)
```

```python
test_q_1()
```

#### answer

```python
q_1(load_data())
```

## part 2

### problem statement:

b#### function def

```python
def get_joltage_2(row: list[int], num_switches: int = 12) -> str:
    """find the largest value on the left edge, then concatenate with the joltage of the remainder
    with one fewer switch

    """
    a = max(row[:len(row) - num_switches + 1])
    i = row.index(a)
    a = str(a)
    if num_switches == 1:
        return a
    else:
        rest = row[i + 1:]
        b = get_joltage_2(rest, num_switches - 1)
        return a + b

assert get_joltage_2(parse_raw_data(data="987654321111111")[0]) == "987654321111"
assert get_joltage_2(parse_raw_data(data="811111111111119")[0]) == "811111111119"
assert get_joltage_2(parse_raw_data(data="234234234234278")[0]) == "434234234278"
assert get_joltage_2(parse_raw_data(data="818181911112111")[0]) == "888911112111"
```

```python
def q_2(data):
    return sum(int(get_joltage_2(row)) for row in parse_raw_data(data=data))
```

#### tests

```python
def test_q_2():
    LOGGER.setLevel(logging.DEBUG)
    assert q_2(test_data) == 3121910778619
    LOGGER.setLevel(logging.INFO)
```

```python
test_q_2()
```

#### answer

```python
q_2(load_data())
```
